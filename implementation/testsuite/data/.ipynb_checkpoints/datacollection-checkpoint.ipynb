```python
import plotly.graph_objects as go
import pandas as pd
colors = {}
colors['Zlnk product based'] = '#1f77b4'
colors['Fixed-point product based'] = '#ff7f0e'
colors['Fixed-point local product based'] = '#2ca02c'
colors['Zlnk fam based - explicit'] = '#d62728'
colors['Zlnk fam based - symbolic'] = '#9467bd'
colors['Incremental pre-solve'] = '#8c564b'
colors['Incremental pre-solve local'] = '#e377c2'
#    '#7f7f7f',  // middle gray
#    '#bcbd22',  // curry yellow-green
#    '#17becf'   // blue-teal
def plot_running_times(names,csvs,sortby,normalize=-1,plottitle="", orderFile="",yaxisrange=None):
    T = pd.read_csv(csvs[0], header=None).sort_values(by=[0])
    T[1] = T[1] / 1000000
    for i in range(1,len(names)):
        F2 = pd.read_csv(csvs[i],header=None).sort_values(by=[0])
        T[i+1] = F2[1] / 1000000
    #T = T.sort_values(by=[sortby+1])
    if orderFile != "":
        F = pd.read_csv(orderFile, header=None).sort_values(by=[0])
        T[len(names)+2] = F[1]
        T = T.sort_values(by=[len(names)+2])

    X=T[0].tolist()

    for i in range(len(X)):
        X[i] = 'Game ' + str(X[i])
    
    table="\\begin{tabular}{|"
    for i in range(len(names)+1):
        table += 'l|'
    table += "}\n\hline\n "
    for n in names:
        table += '& ' + n + ' '
    table += "\\\\ \hline\n"
    for y in range(len(X)):
        table += str(X[y]) + " "
        for x in range(len(names)):
            table += '& '
            val = str(T[x+1].tolist()[y])
            if val == 'nan':
                val = '\\dag'
            else:
                val += ' ms'
            table += val
            table += ' '
        table += "\\\\ \hline\n"
    table += "\\end{tabular}"
    
    if normalize > 0:
        for i in range(1 , len(names)+1):
            if i != normalize:
                T[i] = T[i] / T[normalize]
        T[normalize] = T[normalize] / T[normalize]
    
    D= [None] * len(names)
    for i in range(len(names)):
        if names[i] in colors:
            D[i] = go.Bar(name=names[i], x=X, y=T[i+1].tolist(),marker_color=colors[names[i]])
        else:
            print "Unknown name " + names[i]
    fig = go.Figure(data=D)
    # Change the bar mode
    fig.update_layout(
         legend=go.layout.Legend(
            x=0,
            y=1,
            traceorder="normal",
            font=dict(
                family="sans-serif",
                size=12,
                color="black"
            ),
            bgcolor="LightSteelBlue",
            bordercolor="Black",
            borderwidth=2
        ),
        barmode='group',
#         title=go.layout.Title(
#             text=plottitle
#         ),
    )
    print table
    if yaxisrange != None:
        fig.update_yaxes(range=yaxisrange)
    fig.show()
```
